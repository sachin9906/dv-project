```python
import pandas as pd

# Load the dataset
dataset_path = "/kaggle/input/flipkart-products/flipkart_com-ecommerce_sample.csv"
df = pd.read_csv(dataset_path)

# Display basic information about the dataset
print("Columns and data types:")
print(df.dtypes)

# Display basic statistics
print("\nBasic statistics:")
print(df.describe())

# Display the first few rows of the dataset
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Identify the most popular product categories
top_categories = df['product_category_tree'].value_counts().nlargest(10)

# Plot a bar chart for the top categories
plt.figure(figsize=(12, 8))
top_categories.plot(kind='barh', color='skyblue')
plt.xlabel('Count')
plt.ylabel('Product Category')
plt.title('Top 10 Product Categories')
plt.gca().invert_yaxis()  # Invert y-axis for better visualization
plt.show()

# Identify the most popular products
top_products = df['product_name'].value_counts().nlargest(10)

# Plot a pie chart for the top products
plt.figure(figsize=(10, 10))
plt.pie(top_products, labels=top_products.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Top 10 Popular Products')
plt.show()
```
